```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('factbook.db')

def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
sql_query("Select * from sqlite_master")
```

```python
sql_query("Select * from cities")
```

```python
sql_query("Select * from facts")
```

## Question 1
List all cities where the area is greater than 5000, along with their population.

```python
sql_query('''Select cities.name "City Name", cities.population "City Population" FROM cities
inner join facts
on facts.id = cities.facts_id
where facts.area > 5000''')
```

## Question 2
List the names of capital cities where the death rate is greater than 2% and the
birth rate is less than 20%

```python
sql_query('''Select cities.name "Capital Name" FROM cities
inner join facts
on facts.id = cities.facts_id
where cities.capital == 1 and facts.death_rate > 2 and facts.birth_rate < 20''')
```

## Question 3
List the cities with a population growth rate between 1% and 5%, sorted in desc
by population growth.

```python
sql_query('''Select cities.name "City Name", facts.population_growth "Population Growth Rate" FROM cities
inner join facts
on facts.id = cities.facts_id
where facts.population_growth between 1 and 5 order by 2 desc''')
```

## Question 4
Find countries that start with "A" and do not have any cities listed in the cities
table. The query should return the country name and city name. Using left join

```python
sql_query('''Select facts.name "Country Name", cities.name "City Name" FROM facts
left join cities
on facts.id = cities.facts_id
where facts.name like "A%" and cities.name is null''')
```

## Question 5
List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
sql_query('''Select cities.name "Capital Name", facts.death_rate "Death Rate" FROM cities
left join facts
on facts.id = cities.facts_id
where cities.capital == 1''')
```
