```python
import pandas as pd
```

```python
df = pd.read_csv('compensation data.csv')
df
```

```python
python_user = df[df["python_user"] == True].shape[0]
print("Python users:",python_user)

r_user = df[df["r_user"] == True].shape[0]
print("R users:",r_user)

sql_user = df[df["sql_user"] == True].shape[0]
print("SQL users:", sql_user)
```

```python
min_exp = df["experience_coding"].min()
print("Min experience_coding:", min_exp)

max_exp = df["experience_coding"].max()
print("Max experience_coding:", max_exp)

avg_exp = df["experience_coding"].mean()
print("Avg experience_coding:", avg_exp)
```

```python
min_comp = df["compensation"].min()
print("Min compensation:", min_comp)

max_comp = df["compensation"].max()
print("Max compensation:", max_comp)

avg_comp = df["compensation"].mean()
print("Avg compensation:", avg_comp)
```

```python
df["category"] = "25+ Years"
df.loc[df["experience_coding"] < 26, "category"] = "21-25 Years"
df.loc[df["experience_coding"] < 21, "category"] = "16-20 Years"
df.loc[df["experience_coding"] < 16, "category"] = "11-15 Years"
df.loc[df["experience_coding"] < 11, "category"] = "5-10 Years"
df.loc[df["experience_coding"] < 5, "category"] = "< 5 Years"
df
```

```python
df.groupby("category")["compensation"].mean()
```

```python
df.to_csv("new_file.csv", index=False)
```
